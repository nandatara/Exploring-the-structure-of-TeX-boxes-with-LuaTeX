```python
with open('vagdhenu_doc.tex', 'w') as f:
    f.write(r'''\documentclass[12pt,a4paper]{article}

% ── Packages ──
\usepackage[utf8]{inputenc}
\usepackage[margin=1in]{geometry}
\usepackage{hyperref}
\usepackage{xcolor}
\usepackage{listings}
\usepackage{enumitem}
\usepackage{booktabs}
\usepackage{longtable}
\usepackage{fancyhdr}
\usepackage{titlesec}
\usepackage{tcolorbox}

% ── Hyperlink colors ──
\hypersetup{
    colorlinks=true,
    linkcolor=blue!70!black,
    urlcolor=blue!60!black,
    citecolor=green!50!black,
}

% ── Page headers ──
\pagestyle{fancy}
\fancyhf{}
\fancyhead[L]{Vāgdhenu Sanskrit Chant TTS}
\fancyhead[R]{\thepage}
\renewcommand{\headrulewidth}{0.4pt}

% ── Section formatting ──
\titleformat{\section}{\Large\bfseries\color{purple!70!black}}{\thesection}{1em}{}
\titleformat{\subsection}{\large\bfseries\color{purple!60!black}}{\thesubsection}{1em}{}

% ── Code listing style ──
\lstdefinestyle{pythonstyle}{
    backgroundcolor=\color{gray!5},
    commentstyle=\color{green!50!black},
    keywordstyle=\color{blue},
    stringstyle=\color{red!70!black},
    basicstyle=\ttfamily\small,
    breaklines=true,
    frame=single,
    rulecolor=\color{gray!50},
    numbers=left,
    numberstyle=\tiny\color{gray},
    language=Python,
    showstringspaces=false,
}
\lstset{style=pythonstyle}

\lstdefinestyle{bashstyle}{
    backgroundcolor=\color{gray!5},
    basicstyle=\ttfamily\small,
    breaklines=true,
    frame=single,
    rulecolor=\color{gray!50},
    numbers=none,
    language=bash,
}

% ── Custom environments ──
\newtcolorbox{infobox}[1][]{
    colback=blue!5,
    colframe=blue!40,
    fonttitle=\bfseries,
    title=#1,
    arc=3mm,
}

\newtcolorbox{warningbox}[1][]{
    colback=orange!5,
    colframe=orange!60,
    fonttitle=\bfseries,
    title=#1,
    arc=3mm,
}

\begin{document}

% ════════════════════════════════════════════
% TITLE PAGE
% ════════════════════════════════════════════
\begin{titlepage}
\centering
\vspace*{2cm}

{\Huge\bfseries\color{purple!70!black} Vāgdhenu}\\[0.5cm]
{\Large Sanskrit Chant Text-to-Speech}\\[0.3cm]
{\large Setup and Usage Guide}\\[2cm]

{\large\itshape ``The wish-cow of speech''}\\[0.5cm]
{\small वाग्धेनु}\\[3cm]

\begin{tabular}{ll}
\textbf{Author:} & Prof.\ Prathosh A.\ P., IISc Bengaluru \\
\textbf{License:} & Apache-2.0 \\
\textbf{GitHub:} & \url{https://github.com/prathoshap/vagdhenu} \\
\textbf{Demo:} & \url{https://huggingface.co/spaces/prathoshap/vagdhenu} \\
\end{tabular}

\vfill
{\small \today}
\end{titlepage}

\tableofcontents
\newpage

% ════════════════════════════════════════════
% 1. INTRODUCTION
% ════════════════════════════════════════════
\section{Introduction to Vāgdhenu}

\textbf{Vāgdhenu} (वाग्धेनु, meaning ``the wish-cow of speech'') is an open-source single-speaker Sanskrit chant (\textit{pārāyaṇa}) text-to-speech system developed by Prof.\ Prathosh A.\ P.\ at IISc Bengaluru.

Unlike standard TTS systems that produce flat read-aloud speech, Vāgdhenu \textit{chants} classical ślokas with metrical awareness and traditional melodic contours.

\subsection{Key Features}
\begin{itemize}[noitemsep]
    \item Built on AI4Bharat's \textbf{IndicF5} backbone with NVIDIA \textbf{BigVGAN-v2} vocoder
    \item \textbf{Meter-aware} (vṛtta) synthesis — supports anuṣṭubh, triṣṭubh, gayatri, etc.
    \item Includes reference audio bank for prosody control across 18+ meters
    \item Routes Devanagari text through Kannada orthography to preserve Sanskrit phonology
    \item Handles visarga sandhi, three sibilants, anusvāra variants correctly
\end{itemize}

\subsection{Use Cases}
\begin{itemize}[noitemsep]
    \item Rendering religious texts (Bhagavad Gita, Bhāgavatam, Vedas)
    \item Creating accessible Sanskrit audio content
    \item Language preservation and research
    \item Meditation and music productions
\end{itemize}

\subsection{Real-World Impact}
\begin{itemize}[noitemsep]
    \item \textbf{MBTN YouTube series} — 32 videos totaling over 17 hours
    \item \textbf{Śrīmad Bhāgavatam} — all 16,017 verses rendered with audio app + 31 karaoke videos
\end{itemize}

% ════════════════════════════════════════════
% 2. REQUIREMENTS
% ════════════════════════════════════════════
\section{Requirements}

\subsection{Local Installation}
\begin{itemize}[noitemsep]
    \item Python 3.10
    \item NVIDIA GPU with CUDA 12.1 compatible drivers
    \item Minimum 6GB VRAM recommended (8GB+ for batch processing)
    \item ${\sim}$3GB free storage for model weights
\end{itemize}

\subsection{For Colab Users}
\begin{itemize}[noitemsep]
    \item Google account (free tier sufficient)
    \item No local hardware required
    \item Free GPU access via Colab runtime
    \item ${\sim}$2 hours per session limit (free tier)
\end{itemize}

\subsection{Supported Meters}

\begin{longtable}{ll}
\toprule
\textbf{Meter} & \textbf{Description} \\
\midrule
anuṣṭubh & Standard Gita śloka meter (8 syllables/pāda) \\
triṣṭubh & Longer meter (11 syllables/pāda) \\
upajāti & Mixed meter combining anuṣṭubh and triṣṭubh \\
indravajrā & Triṣṭubh-family meter \\
gadya & Prose meter (used for speaker attributions) \\
vasantatilakā & Ornate meter (14 syllables/pāda) \\
sragdharā & Extended meter (21 syllables/pāda) \\
\bottomrule
\caption{Partial list — reference bank includes 18+ meters}
\end{longtable}

% ════════════════════════════════════════════
% 3. COLAB INFORMATION
% ════════════════════════════════════════════
\section{Information on Colab}

\subsection{Why Use Colab?}
Many users lack local machines with dedicated NVIDIA GPUs. Colab provides:
\begin{itemize}[noitemsep]
    \item Free Tesla T4 GPU (16GB VRAM)
    \item Pre-configured CUDA environment
    \item Internet connectivity for downloading dependencies
    \item No local installation complexity
\end{itemize}

\subsection{Session Limitations}

\begin{warningbox}[Free Tier Limits]
\begin{itemize}[noitemsep]
    \item Maximum 4–12 hours per session (varies)
    \item Session timeout after idle period (${\sim}$90 min)
    \item Files deleted when session ends — download frequently!
    \item May disconnect if you switch tabs or leave browser
\end{itemize}
\end{warningbox}

\subsection{Colab Pro (\$10/month)}
\begin{itemize}[noitemsep]
    \item Longer sessions (up to 24 hours)
    \item Access to better GPUs (A100, L4)
    \item More concurrent sessions
    \item Background execution enabled
\end{itemize}

\subsection{Alternative: Cloud GPU Rental}
\begin{longtable}{lll}
\toprule
\textbf{Platform} & \textbf{Rate} & \textbf{Notes} \\
\midrule
RunPod & \$0.34–0.40/hr & RTX 4090, per-second billing \\
Vast.ai & \$0.15–0.30/hr & Community-hosted, cheapest \\
Lambda Labs & \$0.50/hr & Enterprise-grade reliability \\
\bottomrule
\end{longtable}

% ════════════════════════════════════════════
% 4. INSTALLATION
% ════════════════════════════════════════════
\section{Installation Step by Step}

\begin{infobox}[Prerequisites]
Fresh Colab runtime required. Go to \textbf{Runtime $\rightarrow$ Disconnect and delete session}, then reconnect with GPU enabled (\textbf{Runtime $\rightarrow$ Change runtime type $\rightarrow$ T4 GPU}).
\end{infobox}

\subsection{Step 1: Clone Repositories}

\begin{lstlisting}[style=bashstyle]
import subprocess

def run(cmd):
    print(f">>> {cmd[:80]}...")
    r = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    print("OK" if r.returncode == 0 else f"FAILED: {r.stderr[-300:]}")
    return r

run('git clone https://github.com/prathoshap/vagdhenu.git /content/vagdhenu')
run('git clone https://github.com/NVIDIA/BigVGAN.git /content/BigVGAN')
\end{lstlisting}

\subsection{Step 2: Install Dependencies}

\begin{lstlisting}[style=bashstyle]
# Install vagdhenu deps without touching torch
run('pip install -r /content/vagdhenu/requirements.txt --no-deps')

# Pin numpy (2.x breaks BigVGAN)
run('pip install numpy==1.26.4 --force-reinstall --no-deps')
\end{lstlisting}

\subsection{Step 3: Ensure Matching Torch/TorchVision}

\begin{lstlisting}[style=bashstyle]
run('pip install "torch>=2.4.1,<2.5" torchvision==0.19.1 torchaudio==2.4.1 '
    '--index-url https://download.pytorch.org/whl/cu121')
\end{lstlisting}

Verify versions match:
\begin{lstlisting}[style=bashstyle]
python -c "import torch, torchvision; print(torch.__version__, torchvision.__version__)"
# Both should end in +cu121
\end{lstlisting}

\subsection{Step 4: Download Model Weights}

\begin{lstlisting}[style=bashstyle]
run('cd /content/vagdhenu && bash scripts/setup.sh')
\end{lstlisting}

Downloads the following files to \texttt{/content/vagdhenu/models/}:

\begin{longtable}{ll}
\toprule
\textbf{File} & \textbf{Purpose} \\
\midrule
\texttt{voice\_steer\_ema\_2026-06-17.pt} & Production voice (recommended default) \\
\texttt{voice\_armA\_ema\_2026-06-11.pt} & Fallback option (reference-driven) \\
\texttt{voc\_bigvgan\_EMA\_2026-06-11.pth} & BigVGAN-v2 vocoder (mandatory) \\
\texttt{vocab.txt} & Token vocabulary \\
\bottomrule
\end{longtable}

\subsection{Step 5: Patch the Script}

The GitHub repo has placeholder paths (\texttt{<PROD>}). Fix them:

\begin{lstlisting}[style=pythonstyle]
with open('/content/vagdhenu/src/render_production.py', 'r') as f:
    code = f.read()

code = code.replace('<PROD>', '/content/vagdhenu')
code = code.replace(
    'vocab = glob.glob(os.path.expanduser('
    '"~/.cache/huggingface/hub/models--ai4bharat--IndicF5'
    '/snapshots/*/checkpoints/vocab.txt"))[0]',
    'vocab = "/content/vagdhenu/models/vocab.txt"'
)

with open('/content/vagdhenu/src/render_production_patched.py', 'w') as f:
    f.write(code)
\end{lstlisting}

\subsection{Step 6: Verify Installation}

\begin{lstlisting}[style=pythonstyle]
import subprocess
r = subprocess.run(
    'python -c "import torch,numpy; '
    'print(torch.__version__, numpy.__version__, '
    'torch.cuda.is_available())"',
    shell=True, capture_output=True, text=True
)
print(r.stdout.strip())
# Expected: 2.4.1+cu121 1.26.4 True
\end{lstlisting}

% ════════════════════════════════════════════
% 5. PREPARING JS FILES
% ════════════════════════════════════════════
\section{Preparing .js Files}

\subsection{Data Format}

Input should be structured as a JavaScript object with verse keys like \texttt{"chapter:verse"}:

\begin{lstlisting}[style=bashstyle,language=]
const VERSE_TEXT_OVERRIDES = {
    "1:1": {
        "sa": [
            "speaker attribution",
            "first pada of sloka",
            "second pada of sloka"
        ],
        "iast": [
            "transliteration line 1",
            "transliteration line 2"
        ]
    },
    "1:2": { ... }
};
\end{lstlisting}

\begin{infobox}[Line Structure]
\textbf{Line 0} is speaker attribution (e.g., ``धृतराष्ट्र उवाच''). \\
\textbf{Lines 1+} are the actual śloka padas. \\
Verses with only 2 lines have no speaker attribution.
\end{infobox}

\subsection{Parsing in Colab}

Upload your JS file via drag/drop or Files panel, then:

\begin{lstlisting}[style=pythonstyle]
import json, re

with open('/content/bg/ch1.js', 'r') as f:
    content = f.read()

start = content.index('{')
end = content.rindex('}') + 1
json_str = content[start:end]

# Remove trailing commas (JS allows, JSON doesn't)
json_str = re.sub(r',(\s*[}\]])', r'\1', json_str)

VERSE_DATA = json.loads(json_str)
print(f"Loaded {len(VERSE_DATA)} verses")
\end{lstlisting}

\subsection{Converting Large Files}

Convert JS to pure JSON locally before uploading:

\begin{lstlisting}[style=bashstyle]
# In browser console:
console.log(JSON.stringify(VERSE_TEXT_OVERRIDES, null, 2));
# Copy output, save as verses.json
\end{lstlisting}

Load in Colab:
\begin{lstlisting}[style=pythonstyle]
with open('/content/verses.json', 'r') as f:
    VERSE_DATA = json.load(f)
\end{lstlisting}

% ════════════════════════════════════════════
% 6. RENDERING
% ════════════════════════════════════════════
\section{Rendering}

\subsection{Single Verse Test}

\begin{lstlisting}[style=pythonstyle]
import json, os, subprocess

os.makedirs('/tmp/vagdhenu_temp', exist_ok=True)
os.makedirs('/content/output', exist_ok=True)

test_padas = [
    "verse line 1 in devanagari",
    "verse line 2 in devanagari"
]

with open('/tmp/vagdhenu_temp/test.json', 'w') as f:
    json.dump(test_padas, f, ensure_ascii=False)

env = os.environ.copy()
env['PYTHONPATH'] = '/content/BigVGAN:/content/vagdhenu/src'

result = subprocess.run([
    'python', 'src/render_production_patched.py',
    '--padas', '/tmp/vagdhenu_temp/test.json',
    '--out', '/content/output/test.wav',
    '--meter', 'anushtubh',
    '--bank', '/content/vagdhenu/src/reference_bank/bank.json',
    '--voice', '/content/vagdhenu/models/voice_steer_ema_2026-06-17.pt',
    '--voc', '/content/vagdhenu/models/voc_bigvgan_EMA_2026-06-11.pth',
    '--seed', '42'
], cwd='/content/vagdhenu', env=env,
   stdout=subprocess.PIPE, stderr=subprocess.STDOUT,
   text=True, timeout=300)

if os.path.exists('/content/output/test.wav'):
    size_kb = os.path.getsize('/content/output/test.wav') / 1024
    print(f"Success! Size: {size_kb:.1f} KB")
else:
    print("Failed")
\end{lstlisting}

\subsection{Batch Rendering Script}

\begin{lstlisting}[style=pythonstyle]
import json, os, re, subprocess

TEMP_DIR = '/tmp/vagdhenu_temp'
os.makedirs(TEMP_DIR, exist_ok=True)
os.makedirs('/content/output', exist_ok=True)

env = os.environ.copy()
env['PYTHONPATH'] = '/content/BigVGAN:/content/vagdhenu/src'

# Meter overrides for non-anushtubh verses
METER_OVERRIDES = {
    "3:2": "upajati", "10:25": "upajati",
    "2:31": "indravajra", "4:5": "indravajra",
    "8:12": "indravajra", "11:12": "indravajra",
    "11:13": "indravajra", "11:45": "indravajra",
    "12:2": "indravajra", "12:3": "indravajra",
    "12:4": "indravajra", "12:5": "indravajra",
    "14:20": "indravajra", "15:5": "indravajra",
}

def parse_js(filepath):
    with open(filepath, 'r') as f:
        content = f.read()
    start = content.index('{')
    end = content.rindex('}') + 1
    json_str = content[start:end]
    json_str = re.sub(r',(\s*[}\]])', r'\1', json_str)
    return json.loads(json_str)

def render_verse(padas, output_path, meter, seed=42):
    # Skip if already rendered (resume support)
    if os.path.exists(output_path):
        return True
    with open(f'{TEMP_DIR}/temp.json', 'w') as f:
        json.dump(padas, f, ensure_ascii=False)
    subprocess.run([
        'python', 'src/render_production_patched.py',
        '--padas', f'{TEMP_DIR}/temp.json',
        '--out', output_path,
        '--meter', meter,
        '--bank',
        '/content/vagdhenu/src/reference_bank/bank.json',
        '--voice',
        '/content/vagdhenu/models/voice_steer_ema_2026-06-17.pt',
        '--voc',
        '/content/vagdhenu/models/voc_bigvgan_EMA_2026-06-11.pth',
        '--seed', str(seed)
    ], cwd='/content/vagdhenu', env=env,
       stdout=subprocess.DEVNULL, timeout=120)
    return os.path.exists(output_path)

for ch in range(1, 19):
    filepath = f'/content/bg/ch{ch}.js'
    if not os.path.exists(filepath):
        continue

    verse_data = parse_js(filepath)
    chapter_dir = f'/content/output/ch{ch}'
    os.makedirs(chapter_dir, exist_ok=True)

    for vkey in sorted(verse_data.keys(),
                       key=lambda x: int(x.split(':')[1])):
        vd = verse_data[vkey]
        sa_lines = vd['sa']
        chap, vn = vkey.split(':')

        spk = sa_lines[0] if len(sa_lines) >= 3 else None
        vpada = sa_lines[1:] if spk else sa_lines
        meter = METER_OVERRIDES.get(vkey, 'anushtubh')

        if spk:
            out_spk = f'{chapter_dir}/{chap}_{vn}_speaker.wav'
            if not os.path.exists(out_spk):
                render_verse([spk], out_spk, 'gadya')

        out_v = f'{chapter_dir}/{chap}_{vn}.wav'
        render_verse(vpada, out_v, meter)

    subprocess.run(
        f'cd {chapter_dir} && zip -qr '
        f'/content/ch{ch}_audio.zip .', shell=True)
\end{lstlisting}

\subsection{Estimated Rendering Times}

\begin{longtable}{lll}
\toprule
\textbf{Scope} & \textbf{Verses} & \textbf{Est.\ Time} \\
\midrule
Single verse & 1 & 30–60 seconds \\
Chapter 1 & 47 & ${\sim}$50 minutes \\
Full Gita (Ch.\ 1–18) & 701 & ${\sim}$7 hours \\
\bottomrule
\end{longtable}

\subsection{Meter Overrides for the Bhagavad Gita}

Approximately 645 of 700 verses use anuṣṭubh. The remainder use other meters:

\begin{longtable}{lll}
\toprule
\textbf{Meter} & \textbf{Verses} & \textbf{Vāgdhenu Key} \\
\midrule
upajāti & 3:2, 10:25 (+${\sim}$49 others) & \texttt{upajāti} \\
indravajrā & 2:31, 4:5, 8:12, 11:12, 11:13, 11:45, & \texttt{indravajrā} \\
 & 12:2–5, 14:20, 15:5 & \\
gadya & All speaker attributions & \texttt{gadya} \\
\bottomrule
\end{longtable}

% ════════════════════════════════════════════
% 7. ZIPPING AND DOWNLOADING
% ════════════════════════════════════════════
\section{Zipping and Downloading}

\subsection{Download Individual Files}

From Google Colab Files panel (left sidebar):
\begin{enumerate}[noitemsep]
    \item Navigate to \texttt{/content/output/ch1/}
    \item Right-click individual files
    \item Choose ``Download''
\end{enumerate}

\subsection{Zip Entire Chapter}

\begin{lstlisting}[style=pythonstyle]
# Create compressed archive
subprocess.run(
    'cd /content/output/ch1 && zip -qr '
    '/content/ch1_audio.zip .', shell=True)

# Download via browser
from google.colab import files
files.download('/content/ch1_audio.zip')
\end{lstlisting}

\subsection{Download All Chapters}

\begin{lstlisting}[style=pythonstyle]
from google.colab import files
import os

for ch in range(1, 19):
    zipfile = f'/content/ch{ch}_audio.zip'
    if os.path.exists(zipfile):
        size_mb = os.path.getsize(zipfile) / (1024 * 1024)
        print(f"Downloading ch{ch} ({size_mb:.1f} MB)...")
        files.download(zipfile)
\end{lstlisting}

\subsection{Partial Chapter Recovery}

If a session drops mid-chapter, pack whatever was completed:

\begin{lstlisting}[style=pythonstyle]
import os, subprocess

for ch in range(1, 19):
    chapter_dir = f'/content/output/ch{ch}'
    if os.path.isdir(chapter_dir):
        wav_count = len([
            f for f in os.listdir(chapter_dir)
            if f.endswith('.wav')
        ])
        if wav_count > 0:
            subprocess.run(
                f'cd {chapter_dir} && zip -qr '
                f'/content/ch{ch}_partial.zip .',
                shell=True)
            print(f"ch{ch}: {wav_count} files packed")
\end{lstlisting}

\subsection{File Naming Convention}

Generated files follow the pattern:

\begin{longtable}{ll}
\toprule
\textbf{Pattern} & \textbf{Example} \\
\midrule
\texttt{\{chapter\}\_\{verse\}.wav} & \texttt{1\_01.wav}, \texttt{2\_47.wav} \\
\texttt{\{chapter\}\_\{verse\}\_speaker.wav} & \texttt{1\_01\_speaker.wav} \\
\texttt{ch\{N\}\_audio.zip} & \texttt{ch1\_audio.zip} \\
\bottomrule
\end{longtable}

% ════════════════════════════════════════════
% 8. TROUBLESHOOTING
% ════════════════════════════════════════════
\section{Troubleshooting}

\begin{longtable}{p{4.5cm}p{4cm}p{6cm}}
\toprule
\textbf{Error} & \textbf{Likely Cause} & \textbf{Fix} \\
\midrule
\texttt{ncclCommResume} undefined symbol & NCCL library conflict & Restart runtime; reinstall matching torch/torchvision \\
\texttt{torchvision::nms does not exist} & PyTorch/torchvision mismatch & Reinstall both from same index URL \\
\texttt{No module named 'bigvgan'} & PYTHONPATH not set & Add BigVGAN path to subprocess env \\
\texttt{numpy.strings} import error & NumPy 2.x installed & \texttt{pip install numpy==1.26.4} \\
Session timeout & Idle or memory limit & Reduce batch size; download frequently \\
Out of memory (OOM) & Too many concurrent renders & Process one verse at a time \\
\bottomrule
\end{longtable}

% ════════════════════════════════════════════
% 9. REFERENCES
% ════════════════════════════════════════════
\section{References and Credits}

\subsection{Original Project}
\begin{itemize}[noitemsep]
    \item Author: Prof.\ Prathosh A.\ P., IISc Bengaluru
    \item License: Apache-2.0
    \item GitHub: \url{https://github.com/prathoshap/vagdhenu}
\end{itemize}

\subsection{Technology Stack}
\begin{itemize}[noitemsep]
    \item F5-TTS framework (flow-matching DiT architecture)
    \item AI4Bharat IndicF5 base model
    \item NVIDIA BigVGAN-v2 vocoder
    \item Half-reference rule for prosody control
\end{itemize}

\subsection{Acknowledgements}
\begin{itemize}[noitemsep]
    \item LatentForce (infrastructure support)
    \item AI4Bharat initiative
    \item Open-source community for F5-TTS and BigVGAN
\end{itemize}

% ════════════════════════════════════════════
% 10. LICENSE
% ════════════════════════════════════════════
\section{License}

Vāgdhenu project is licensed under Apache-2.0. See the original repository for full terms.

\begin{warningbox}[Usage Notes]
\begin{itemize}[noitemsep]
    \item Single-speaker constraint: voice belongs to Prof.\ Prathosh
    \item Do not impersonate the original speaker
    \item Cite the Vāgdhenu project in derivative works
\end{itemize}
\end{warningbox}

\vfill
\begin{center}
\small\itshape Document generated July 2026 \\
Vāgdhenu Sanskrit Chant TTS — Setup Guide
\end{center}

\end{document}''')

!pdflatex -interaction=nonstopmode vagdhenu_doc.tex
!pdflatex -interaction=nonstopmode vagdhenu_doc.tex

from google.colab import files
files.download('vagdhenu_doc.pdf')
```
